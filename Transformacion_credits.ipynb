```python
import json
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/cesua/OneDrive/Documentos/Henry/Proyecto_Individual_1/Datasets/credits.csv")
df.head(3)
```

```python
# Obtener las columnas actuales del DataFrame
columnas_actuales = df.columns.tolist()

# Definir el nuevo orden de las columnas
nuevo_orden = ['id', 'cast', 'crew']

# Reordenar las columnas del DataFrame
df = df[nuevo_orden]
```

```python
def desanidar_credits(credits): 
    creditos_director = []
    creditos_nombre = []
    creditos_id = []
    for creditos in credits:
        creditos_director.append(creditos['job'])
        creditos_nombre.append(creditos['name'])
        creditos_id.append(creditos['id'])
    return creditos_director, creditos_nombre, creditos_id

creditos_desanidados = df['crew'].apply(lambda x: desanidar_credits(eval(x)))

nuevo_df = pd.DataFrame(creditos_desanidados.tolist(), columns=['job', 'name','id'])

nuevo_df = nuevo_df.apply(pd.Series.explode)

nuevo_df = nuevo_df.reset_index(drop=True)
```

```python
# Unir el DataFrame original con las columnas 'job' y 'name' del nuevo DataFrame
df = df.join(nuevo_df[['job', 'name']])
```
